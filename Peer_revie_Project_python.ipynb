```python
# Question 1 - Extracting Tesla Stock Data Using yfinance - 2 Points
import yfinance as yf

tesla = yf.Ticker("TSLA")

tesla_data = tesla.history(period="max")

tesla_data.reset_index(inplace=True)

tesla_data.head()
```

```python
# Question 2 - Extracting Tesla Revenue Data Using Webscraping - 1 Points
import requests
import pandas as pd
from bs4 import BeautifulSoup
import warnings

warnings.filterwarnings("ignore", category=FutureWarning)

url = 'https://companiesmarketcap.com/tesla/revenue/'

response = requests.get(url).text

beautiful_soup = BeautifulSoup(response, "html.parser")

html_string = str(beautiful_soup)

tesla_revenue = pd.read_html(html_string, match="Revenue", flavor='bs4')[0]

print(tesla_revenue.head())
```

```python
# Question 3 - Extracting GameStop Stock Data Using yfinance - 2 Points

gme = yf.Ticker("TSLA")

gme_data = gme.history(period="max")

gme_data.reset_index(inplace=True)

gme_data.head()
```

```python
# Question 4 - Extracting GameStop Revenue Data Using Webscraping - 1 Points

url = 'https://companiesmarketcap.com/gamestop/revenue/'

response = requests.get(url).text

beautiful_soup = BeautifulSoup(response, "html.parser")

html_string = str(beautiful_soup)

gme_revenue = pd.read_html(html_string, match="Revenue", flavor='bs4')[0]

print(gme_revenue.head())
```

```python
# Question 5 - Tesla Stock and Revenue Dashboard - 2 Points

def convert_to_numeric(value):
    value = value.replace('$', '').replace(',', '').strip()
    if 'B' in value:
        return float(value.replace('B', '').strip()) * 1e9  # Convert billions to full numeric value
    elif 'T' in value:
        return float(value.replace('T', '').strip()) * 1e12  # Convert trillions to full numeric value
    else:
        return float(value)

rev_url = 'https://companiesmarketcap.com/tesla/revenue/'
response_1 = requests.get(rev_url).text
tesla_revenue = pd.read_html(response_1, match="Revenue")[0]

print(tesla_revenue.head())

mc_url = 'https://companiesmarketcap.com/tesla/marketcap/'
response_2 = requests.get(mc_url).text

soup_mc = BeautifulSoup(response_2, 'html.parser')

table_mc = soup_mc.find('table')

tesla_market_cap = pd.read_html(str(table_mc))[0]

print(tesla_market_cap.head())

tesla_revenue_processed = tesla_revenue[['Year', 'Revenue', 'Change']]
tesla_revenue_processed.columns = ['Year', 'Revenue', 'Revenue Change']

tesla_market_cap_processed = tesla_market_cap[['Year', 'Market cap', 'Change']]
tesla_market_cap_processed.columns = ['Year', 'Market cap', 'Market Cap Change']

tesla_revenue_processed['Year'] = tesla_revenue_processed['Year'].astype(str)
tesla_market_cap_processed['Year'] = tesla_market_cap_processed['Year'].astype(str)

tesla_revenue_processed['Revenue'] = tesla_revenue_processed['Revenue'].apply(convert_to_numeric)
tesla_market_cap_processed['Market cap'] = tesla_market_cap_processed['Market cap'].apply(convert_to_numeric)

tesla_data = pd.merge(tesla_revenue_processed, tesla_market_cap_processed, on='Year', how='inner')
print(tesla_data.head())

tesla_data.to_csv('tesla_stock_and_revenue.csv', index=False)
```
